## 1.2 Détection et traitement des doublons

L'analyse initiale a révélé **4874 villes pour 4816 noms uniques** → environ 58 doublons potentiels à investiguer.
Un nom de ville dupliqué n'est pas forcément une erreur (ex: *Springfield* existe dans plusieurs États américains).

```python
# ── Détection des doublons par (ville, pays) ───────────────
dup_mask = df.duplicated(subset=['city','country'], keep=False)
n_dup_true = dup_mask.sum()

# ── Détection des doublons par nom de ville seul ───────────
dup_name = df['city'].duplicated(keep=False)
n_dup_names = dup_name.sum()

print(f"🔍 Doublons STRICTS (ville + pays identiques) : {n_dup_true}")
print(f"🔍 Noms de villes dupliqués (pays différents) : {n_dup_names - n_dup_true}")
print(f"\n→ Les noms dupliqués mais dans des pays différents sont LÉGITIMES")
print(f"  (ex: Springfield existe dans plusieurs pays)\n")

if n_dup_true > 0:
    print("📋 Aperçu des vrais doublons :")
    display(df[dup_mask].sort_values(['country','city']).head(10)[['city','country','avg_net_salary','rent_1br_center']])

# ── Traitement : moyenne des valeurs pour les vrais doublons ──
if n_dup_true > 0:
    cols_num = df.select_dtypes(include='number').columns.tolist()
    df = df.groupby(['city','country','continent'], as_index=False)[cols_num].mean()
    print(f"\n✅ Après fusion : {len(df)} lignes (doublons agrégés par moyenne)")
else:
    print("\n✅ Aucun doublon strict à traiter")
```

## 1.3 Détection d'incohérences

Certaines villes peuvent présenter des valeurs contradictoires : salaire élevé mais coût de vie nul, ou inversement.
On flag ces incohérences sans les supprimer automatiquement (elles peuvent refléter une réalité ou un problème de collecte).

```python
# ── Détection d'incohérences économiques ──────────────────
incoherences = []

# Règle 1 : salaire > 2000 USD mais repas < 1 USD (suspect)
r1 = (df['avg_net_salary'] > 2000) & (df['meal_cheap_restaurant'] < 1)
incoherences.append(('Salaire élevé + repas quasi-gratuit', r1.sum()))

# Règle 2 : prix m² centre < 100 USD (probable erreur)
r2 = df['sqm_buy_center'] < 100
incoherences.append(('Prix m² centre anormalement bas (<100$)', r2.sum()))

# Règle 3 : loyer 3ch < loyer 1ch (incohérent)
r3 = df['rent_3br_center'] < df['rent_1br_center']
incoherences.append(('Loyer 3ch < loyer 1ch (centre)', r3.sum()))

# Règle 4 : bière importée < bière locale (rare mais possible)
r4 = df['imported_beer_restaurant'] < df['domestic_beer_restaurant'] * 0.8
incoherences.append(('Bière importée << bière locale', r4.sum()))

# Règle 5 : salaire < 50 USD/mois (erreur probable)
r5 = df['avg_net_salary'] < 50
incoherences.append(('Salaire < 50 USD (probable erreur)', r5.sum()))

# ── Affichage ──────────────────────────────────────────────
print("🔎 INCOHÉRENCES DÉTECTÉES\n")
for label, n in incoherences:
    print(f"   {label:45s} : {n:3d} cas")

# ── Flag dans le dataframe ─────────────────────────────────
df['flag_incoherent'] = (r1 | r3 | r5).astype(int)
print(f"\n⚠️  Total villes flaggées comme suspectes : {df['flag_incoherent'].sum()}")
print("→ Elles restent dans le dataset mais seront surveillées dans les analyses suivantes")
```

## 2.4 Matrice de nullité — Corrélation des valeurs manquantes

Objectif : vérifier si les valeurs manquantes sont **corrélées entre variables**.
- Si les NaN de deux variables apparaissent **ensemble** → mécanisme MAR (Missing At Random conditionnel)
- Si les NaN sont **indépendants** → mécanisme MCAR
- Cela confirme notre diagnostic de la section 3.1

## 3.3 Imputation des Valeurs Manquantes — Baseline médiane globale

```python
# ── Colonnes numériques uniquement ────────────────────────
cols_num = df.select_dtypes(include="number").columns.tolist()
cols_num = [c for c in cols_num if c != "data_quality"]

# ── Imputation par médiane ─────────────────────────────────
df_clean = df.copy()
for col in cols_num:
    median_val = df_clean[col].median()
    df_clean[col] = df_clean[col].fillna(median_val)

# ── Vérification ──────────────────────────────────────────
nan_restant = df_clean[cols_num].isnull().sum().sum()
print(f"✅ Valeurs manquantes après imputation : {nan_restant}")
print(f"📊 Colonnes imputées : {len(cols_num)}")
```

La médiane a été utilisée en raison de l’asymétrie de la distribution et de sa robustesse face aux valeurs extrêmes.

## Imputation avancée — Comparaison des approches

L'imputation par **médiane globale** utilisée plus haut est simple mais **ignore le contexte national** : le prix du pain au Japon ≠ au Kenya.
On compare ici **3 stratégies** pour évaluer l'impact du choix d'imputation :

| Méthode | Principe | Avantage |
|---------|----------|----------|
| **Médiane globale** (baseline) | Une valeur mondiale par variable | Simple, rapide |
| **Médiane par pays** | Une valeur locale par pays | Respecte le contexte économique |
| **MICE (Iterative Imputer)** | Prédit chaque NaN via régression sur autres variables | Exploite les corrélations |


```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.linear_model import BayesianRidge
import warnings
warnings.filterwarnings('ignore')

# ── Colonnes numériques à imputer ──────────────────────────
cols_num = df.select_dtypes(include='number').columns.tolist()
cols_num = [c for c in cols_num if c not in ['data_quality','flag_incoherent']]

# ── Stratégie 1 : médiane globale (BASELINE, déjà fait) ───
# df_clean existe déjà depuis la cellule précédente

# ── Stratégie 2 : médiane par pays (fallback médiane globale) ──
df_country = df.copy()
for col in cols_num:
    df_country[col] = df_country.groupby('country')[col].transform(
        lambda s: s.fillna(s.median())
    )
    # Fallback global si pays entier vide
    df_country[col] = df_country[col].fillna(df[col].median())

# ── Stratégie 3 : MICE (Iterative Imputer) ─────────────────
df_mice = df.copy()
imputer = IterativeImputer(
    estimator=BayesianRidge(),
    max_iter=10, random_state=42,
    min_value=0  # pas de prix négatifs
)
df_mice[cols_num] = imputer.fit_transform(df_mice[cols_num])

# ── Vérification ───────────────────────────────────────────
print("✅ IMPUTATION — 3 DATASETS GÉNÉRÉS\n")
print(f"{'Méthode':<25} {'NaN restants':<15} {'Taille':<10}")
print("─" * 50)
print(f"{'Médiane globale':<25} {df_clean[cols_num].isnull().sum().sum():<15} {len(df_clean)}")
print(f"{'Médiane par pays':<25} {df_country[cols_num].isnull().sum().sum():<15} {len(df_country)}")
print(f"{'MICE (régression)':<25} {df_mice[cols_num].isnull().sum().sum():<15} {len(df_mice)}")
```

### Comparaison des 3 imputations sur variables clés

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

vars_test = ['avg_net_salary', 'rent_1br_center']
labels_test = ['Salaire net', 'Loyer 1ch centre']

colors_m = {'Médiane globale': '#3498db',
            'Médiane pays':    '#27ae60',
            'MICE':            '#e74c3c'}

fig = make_subplots(rows=1, cols=2, subplot_titles=labels_test,
                    horizontal_spacing=0.12)

for (col, lab), (r, c) in zip(zip(vars_test, labels_test), [(1,1),(1,2)]):
    p95 = df_mice[col].quantile(0.95)
    for method, d in [('Médiane globale', df_clean),
                      ('Médiane pays',    df_country),
                      ('MICE',            df_mice)]:
        data_filt = d[col][d[col] <= p95]
        fig.add_trace(
            go.Histogram(x=data_filt, name=method,
                        marker=dict(color=colors_m[method]),
                        opacity=0.55, nbinsx=50,
                        showlegend=(r==1 and c==1),
                        legendgroup=method),
            row=r, col=c
        )
        fig.add_vline(x=d[col].median(),
                     line=dict(color=colors_m[method], width=2, dash='dash'),
                     row=r, col=c)

fig.update_layout(
    title=dict(text='📊 Comparaison des 3 stratégies d\'imputation',
               x=0.5, font=dict(size=14, color='#2c3e50')),
    barmode='overlay', template='plotly_white',
    height=450, margin=dict(t=80, b=40),
    legend=dict(x=1.02, y=1)
)
fig.update_xaxes(title='Valeur (USD)')
fig.update_yaxes(title='Nombre de villes')
fig.show()
```

### 📝 Interprétation — Comparaison des 3 stratégies d'imputation

### 📊 Comment lire ce graphique
- Chaque **barre** = nombre de villes ayant cette valeur
- Les **3 couleurs** représentent les 3 méthodes d'imputation
- Les **lignes verticales pointillées** = médiane de chaque méthode

### 📊 Observations

- 🔵 **Médiane globale** : crée un **pic artificiel massif**
  autour de 1300$ (salaire) et 520$ (loyer)
  → toutes les villes sans données reçoivent la même valeur → irréaliste

- 🟢 **Médiane par pays** : pic vert isolé autour de **3500-4000$**
  (salaire) et **1400$** (loyer) → sur-représente certains pays
  → meilleure que la globale mais encore imparfaite

- 🔴 **MICE** : distribution la plus **naturelle et continue**
  → pas de pic artificiel, reflète mieux la réalité

### 🎯 Choix final — MICE
- MICE produit la distribution la plus **réaliste**
- Il exploite les **corrélations entre variables** pour reconstruire
  les valeurs manquantes
- → **df_imputed = df_mice** pour la suite de l'analyse

## 3.4 Log-transformation des valeurs extrêmes

**Principe clé** : un prix élevé à Monaco ou Zurich n'est **pas une erreur**,
c'est la réalité économique. On ne **supprime pas** les outliers, on **réduit
leur impact** via la log-transformation :

$$df\_log = \log(1 + x)$$

→ `log(1+x)` gère les zéros et réduit l'asymétrie des distributions

```python
from scipy.stats import skew
import numpy as np

# ── Log-transformation ─────────────────────────────────────
df_log = df_imputed.copy()
for col in cols_num:
    df_log[col] = np.log1p(df_log[col])

# ── Comparaison de l'asymétrie (skewness) ──────────────────
vars_skew_check = ['internet_monthly', 'sqm_buy_center', 'avg_net_salary',
                   'rent_1br_center', 'intl_school_yearly']

skew_df = pd.DataFrame({
    'Original'      : [skew(df_imputed[c].dropna()) for c in vars_skew_check],
    'Log-transformé': [skew(df_log[c].dropna())     for c in vars_skew_check],
}, index=vars_skew_check).round(2)

print("📊 RÉDUCTION DE L'ASYMÉTRIE (skewness)")
print("    (valeurs proches de 0 = distribution symétrique)\n")
print(skew_df.to_string())
```

```python
col = 'internet_monthly'

fig = make_subplots(rows=1, cols=2,
                    subplot_titles=('Original (asymétrique)',
                                    'Log-transformé (log(1+x))'),
                    horizontal_spacing=0.1)

fig.add_trace(go.Histogram(x=df_imputed[col], nbinsx=60,
                           marker=dict(color='#e74c3c'), showlegend=False), row=1, col=1)
fig.add_trace(go.Histogram(x=df_log[col], nbinsx=60,
                           marker=dict(color='#27ae60'), showlegend=False), row=1, col=2)

fig.update_layout(
    title=dict(text=f'Effet de la log-transformation — {col}', x=0.5,
               font=dict(size=13, color='#2c3e50')),
    height=400, template='plotly_white', showlegend=False
)
fig.show()
```

### 📝 Interprétation — Effet de la log-transformation

### 📊 Original (rouge)
- Distribution **fortement asymétrique** → pic massif entre 0 et 50\$
- La majorité des villes ont un internet < 100\$/mois
- Quelques valeurs extrêmes s'étirent jusqu'à 600$ → queue très longue à droite

### 📊 Log-transformé (vert)
- Distribution **quasi-symétrique** centrée entre 3 et 5 (log scale)
- Forme en **cloche** → beaucoup plus proche d'une distribution normale
- Les outliers extrêmes sont **ramenés à une échelle raisonnable**

### ⚠️ Observation clé
- La log-transformation **ne supprime pas** les outliers, elle réduit leur impact
- → `df_log` sera utilisé pour l'**ACP et le clustering**
- → `df_imputed` reste le dataset de référence pour les
  **analyses descriptives** en USD

## 3.5 Normalisation — Préparation pour les analyses multivariées

Les variables ont des **échelles très différentes** :
- Prix du café : ~2 USD
- Prix du m² centre : ~3200 USD

Sans normalisation, les variables à grande échelle **dominent** les analyses (ACP, clustering, distances).
On prépare une **version standardisée** du dataset pour les analyses multivariées.

```python
from sklearn.preprocessing import StandardScaler, RobustScaler

# ── Scaler standard (z-score) sur df_log ──────────────────
scaler_std = StandardScaler()
X_std = scaler_std.fit_transform(df_log[cols_num])

# ── Scaler robuste (médiane + IQR) sur df_imputed ──────────
scaler_rob = RobustScaler()
X_rob = scaler_rob.fit_transform(df_imputed[cols_num])

# ── Enregistrement comme DataFrames ────────────────────────
df_std = pd.DataFrame(X_std, columns=cols_num, index=df_log.index)
df_std['city']      = df_log['city'].values
df_std['country']   = df_log['country'].values
df_std['continent'] = df_log['continent'].values

df_rob = pd.DataFrame(X_rob, columns=cols_num, index=df_imputed.index)
df_rob['city']      = df_imputed['city'].values
df_rob['country']   = df_imputed['country'].values
df_rob['continent'] = df_imputed['continent'].values

print("✅ DATASETS NORMALISÉS PRÊTS\n")
print(f"{'Dataset':<15} {'Shape':<15} {'Usage recommandé'}")
print("─" * 65)
print(f"{'df_imputed':<15} {str(df_imputed.shape):<15} Analyses descriptives (USD)")
print(f"{'df_log':<15} {str(df_log.shape):<15} ACP, clustering (log)")
print(f"{'df_std':<15} {str(df_std.shape):<15} Analyses multivariées (z-score sur log)")
print(f"{'df_rob':<15} {str(df_rob.shape):<15} Analyses robustes (RobustScaler)")

# ── Vérification : moyennes et écarts-types après scaling ──
print(f"\n📊 Après StandardScaler : moyenne ≈ {X_std.mean():.2e}, std = {X_std.std():.2f}")
print(f"📊 Après RobustScaler  : médiane ≈ {np.median(X_rob):.2f}, IQR normalisé")
```

## 4.6 Indice Big Mac maison — Pouvoir d'achat relatif

L'**indice Big Mac** (The Economist, 1986) est un proxy mondial du pouvoir d'achat.
On calcule ici un indice équivalent :
$$\text{Big Mac Index} = \frac{\text{Salaire net mensuel}}{\text{Prix McMeal}} = \text{nombre de McMeal achetables}$$


```python
# ── Calcul Big Mac Index par pays ─────────────────────────
bm = df_imputed.groupby(['country','continent']).agg(
    salaire  =('avg_net_salary','median'),
    mcmeal   =('mcmeal','median'),
    cappuccino=('cappuccino','median'),
    n_villes =('city','count')
).reset_index()

bm = bm[bm['n_villes'] >= 2]
bm['bigmac_index'] = (bm['salaire'] / bm['mcmeal']).round(0)
bm['cafe_index']   = (bm['salaire'] / bm['cappuccino']).round(0)

# ── Top 15 / Flop 15 ──────────────────────────────────────
top15 = bm.nlargest(15, 'bigmac_index').sort_values('bigmac_index')
flop15 = bm.nsmallest(15, 'bigmac_index').sort_values('bigmac_index', ascending=False)

fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=(
        '🍔 TOP 15 — Plus fort pouvoir d\'achat (Big Mac)',
        '🍔 FLOP 15 — Plus faible pouvoir d\'achat'
    ),
    horizontal_spacing=0.18
)

fig.add_trace(
    go.Bar(y=top15['country'], x=top15['bigmac_index'],
           orientation='h', marker=dict(color='#27ae60'),
           text=[f'{int(v)} 🍔' for v in top15['bigmac_index']],
           textposition='outside', showlegend=False),
    row=1, col=1
)

fig.add_trace(
    go.Bar(y=flop15['country'], x=flop15['bigmac_index'],
           orientation='h', marker=dict(color='#e74c3c'),
           text=[f'{int(v)} 🍔' for v in flop15['bigmac_index']],
           textposition='outside', showlegend=False),
    row=1, col=2
)

fig.update_layout(
    title=dict(text='🍔 Big Mac Index — Nombre de McMeal achetables avec le salaire médian',
               x=0.5, font=dict(size=14, color='#2c3e50')),
    height=600, template='plotly_white', margin=dict(t=80, b=40)
)
fig.show()
```

## 4.7 Pouvoir d'achat relatif — Combien un salaire médian permet-il d'acheter ?

Synthèse : pour un **panier de biens standards**, combien d'unités par mois ?

```python
# ── Biens du panier ────────────────────────────────────────
panier = {
    '🍔 McMeal'        : 'mcmeal',
    '☕ Cappuccino'     : 'cappuccino',
    '🍺 Bière locale'   : 'domestic_beer_restaurant',
    '🍽️ Repas simple'   : 'meal_cheap_restaurant',
    '🥛 Lait (1L)'      : 'milk_1l',
    '🍞 Pain (500g)'    : 'bread_500g',
    '🍗 Poulet (1kg)'   : 'chicken_1kg',
    '🎬 Ticket cinéma'  : 'cinema_seat',
}

# ── Sélection de pays représentatifs ──────────────────────
pays_focus = ['Switzerland','United States','Germany','Japan',
              'Turkey','Tunisia','India','Nigeria']

pouvoir = []
for pays in pays_focus:
    d = df_imputed[df_imputed['country']==pays]
    if len(d) == 0:
        continue
    sal = d['avg_net_salary'].median()
    row = {'Pays': pays, 'Salaire médian (USD)': round(sal,0)}
    for label, col in panier.items():
        prix = d[col].median()
        row[label] = round(sal/prix, 0) if prix > 0 else 0
    pouvoir.append(row)

pouvoir_df = pd.DataFrame(pouvoir).set_index('Pays')
print("💰 POUVOIR D'ACHAT RELATIF — Nombre d'unités achetables / mois\n")
print(pouvoir_df.to_string())
```

## 5.3 Coefficient de variation — Comparaison de la dispersion

Le **coefficient de variation** (CV = σ/μ) permet de comparer la dispersion entre variables d'échelles différentes.
- CV < 0.3 : faible dispersion (prix stables)
- CV 0.3–0.7 : dispersion modérée
- CV > 0.7 : forte dispersion (prix hétérogènes à l'échelle mondiale)

```python
import plotly.express as px
import pandas as pd

# ── Variables à analyser ───────────────────────────────────
vars_cv = {
    'Salaire net'        : 'avg_net_salary',
    'Loyer centre'       : 'rent_1br_center',
    'Loyer extérieur'    : 'rent_1br_outside',
    'Prix m² centre'     : 'sqm_buy_center',
    'Repas bon marché'   : 'meal_cheap_restaurant',
    'Repas mid-range'    : 'meal_midrange_2people',
    'McMeal'             : 'mcmeal',
    'Essence 1L'         : 'gasoline_1l',
    'Internet'           : 'internet_monthly',
    'Cappuccino'         : 'cappuccino',
    'Cinéma'             : 'cinema_seat',
    'Transport mensuel'  : 'transport_monthly_pass',
    'École internationale': 'intl_school_yearly',
    'Charges 85m²'       : 'utilities_85m2',
    'Fitness'            : 'fitness_club_monthly',
}

# ── Calcul du CV ───────────────────────────────────────────
cv_data = []
for label, col in vars_cv.items():
    d = df_imputed[col].dropna()
    cv_data.append({
        'Variable': label,
        'Moyenne' : d.mean(),
        'Std'     : d.std(),
        'CV'      : d.std()/d.mean()
    })

cv_df = pd.DataFrame(cv_data).sort_values('CV')

# ── Visualisation ──────────────────────────────────────────
colors = ['#27ae60' if v < 0.3 else '#f39c12' if v < 0.7 else '#e74c3c'
          for v in cv_df['CV']]

fig = go.Figure(go.Bar(
    y=cv_df['Variable'], x=cv_df['CV'],
    orientation='h', marker=dict(color=colors),
    text=[f'{v:.2f}' for v in cv_df['CV']],
    textposition='outside'
))

fig.add_vline(x=0.3, line_dash='dot', line_color='#27ae60',
              annotation_text='Seuil 0.3', annotation_position='top')
fig.add_vline(x=0.7, line_dash='dot', line_color='#e74c3c',
              annotation_text='Seuil 0.7', annotation_position='top')

fig.update_layout(
    title=dict(text='📏 Coefficient de variation par variable', x=0.5,
               font=dict(size=14, color='#2c3e50')),
    xaxis_title='CV (écart-type / moyenne)',
    height=600, template='plotly_white',
    margin=dict(t=80, b=40, l=160)
)
fig.show()
```

## 6.3 K-Means sur ACP — Clustering des villes

### 6.3.1 Choix du nombre de clusters — Méthode du coude

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# ── Données ACP (PC1 à PC5) ───────────────────────────────
X_km = X_pca[:, :5]

# ── Méthode du coude ──────────────────────────────────────
inertias    = []
sil_scores  = []
K_range     = range(2, 10)

for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X_km)
    inertias.append(km.inertia_)
    sil_scores.append(silhouette_score(X_km, km.labels_))

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# ── Courbe inertie ────────────────────────────────────────
axes[0].plot(K_range, inertias, "o-",
             color="#2c7bb6", linewidth=2.5, markersize=7)
for k, v in zip(K_range, inertias):
    axes[0].annotate(f"{v:.0f}",
                     xy=(k, v), xytext=(0, 8),
                     textcoords="offset points",
                     ha="center", fontsize=7.5,
                     color="#2c3e50")
axes[0].set_xlabel("Nombre de clusters (k)", fontsize=10)
axes[0].set_ylabel("Inertie", fontsize=10)
axes[0].set_title("Méthode du Coude — Inertie",
                  fontweight="bold", pad=10)
axes[0].set_xticks(list(K_range))

# ── Silhouette ────────────────────────────────────────────
axes[1].plot(K_range, sil_scores, "o-",
             color="#e74c3c", linewidth=2.5, markersize=7)
for k, v in zip(K_range, sil_scores):
    axes[1].annotate(f"{v:.2f}",
                     xy=(k, v), xytext=(0, 8),
                     textcoords="offset points",
                     ha="center", fontsize=7.5,
                     color="#2c3e50")
axes[1].set_xlabel("Nombre de clusters (k)", fontsize=10)
axes[1].set_ylabel("Score Silhouette", fontsize=10)
axes[1].set_title("Score Silhouette — Qualité des clusters",
                  fontweight="bold", pad=10)
axes[1].set_xticks(list(K_range))

plt.suptitle("K-Means — Choix optimal du nombre de clusters",
             fontweight="bold", fontsize=11, y=1.02)
plt.tight_layout()
plt.show()

print("\n📊 Score Silhouette par k :")
for k, s in zip(K_range, sil_scores):
    print(f"   k={k} : {s:.4f}")
```

## 📝 Interprétation

### 📊 Méthode du coude
- L’inertie diminue fortement jusqu’à k = 4  
  → les clusters deviennent progressivement plus pertinents  
- Après k = 4, la baisse ralentit nettement  
  → les gains supplémentaires deviennent faibles  
- Le coude observé à k = 4 indique donc une structuration naturelle des données  

### 📊 Score de silhouette
- k = 2 → score = 0.43  
  → très bonne séparation, mais segmentation trop simpliste  
- k = 3 → score = 0.35  
  → bon équilibre entre séparation et interprétabilité  
- k ≥ 5 → chute du score  
  → clusters moins cohérents et plus difficiles à distinguer  

### ⚠️ Observation clé
- k = 4 représente le meilleur compromis  
  → séparation correcte des groupes  
  → tout en conservant une richesse d’analyse suffisante  
- Ce choix permet d’identifier des **profils de villes distincts et interprétables**
